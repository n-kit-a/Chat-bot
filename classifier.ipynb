```
import sister
import pickle
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
embedder = sister.MeanEmbedding(lang="en")
```

```
df = pd.read_csv('/content/Dataset_eng.csv')
```

```
df
```

```
n = []
for index, row in df.iterrows():
  n.append(embedder(df.iloc[index]['Object']))
```

```
df_vec = pd.DataFrame(n)
```

```
df_vec
```

```
df_vec1 = pd.concat([df, df_vec], axis = 1)
```

```
df_vec1
```

```
df_vec1.drop('Object', axis=1)
```

```
df_vec1 = df_vec1.sample(frac=1) 
y = df_vec1[' Class']
X = df_vec1.drop(columns=[' Class', 'Object'])
```

```
model = KNeighborsClassifier(n_neighbors=2)
model.fit(X, y)
```

```
# save the model to disk
filename = 'finalized_model.sav'
pickle.dump(model, open(filename, 'wb'))
```

```
# load the model from disk
loaded_model = pickle.load(open(filename, 'rb'))
```

```
k = embedder('Hi bro')
k = pd.DataFrame(k)
loaded_model.predict(k.T)
```
